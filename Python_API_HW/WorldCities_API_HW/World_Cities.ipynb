```python
# Dependencies
import csv
import matplotlib.pyplot as plt
import requests
import pandas as pd
import json
```

```python
world_cities = pd.read_csv("worldcities.csv", dtype = str)
world_cities.head()
```

```python
selected_cities = world_cities.sample(n=600)
```

```python
params = {
    "key": "25bc90a1196e6f153eece0bc0b0fc9eb"
}
# Counter
row_count = 0

for index, row in selected_cities.iterrows():   
    params['City'] = row['City']
    base_url = "http://api.openweathermap.org/data/2.5/weather?"

    # Print log to ensure loop is working correctly
    print("Processing Record Number " + str(row_count))
    row_count += 1

    # Run requests to grab the JSON at the requested URL
    zip_location = requests.get(base_url, params=params)
    print(zip_location.url)
```

```python
allcities_df = pd.DataFrame({"City Name": selected_cities["City"]})
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"
api_key = "25bc90a1196e6f153eece0bc0b0fc9eb"
# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
query_url
```

```python
cities_list = allcities_df
for index, row in cities_list.iterrows():      
    target_city = row["City Name"]
    city_response = requests.get(query_url + target_city).json()
    #print(json.dumps(city_response, indent = 4, sort_keys = True))
    try:
        city_name = city_response['name']    
        city_country = city_response['sys']['country']
        city_lon = city_response['coord']['lon']
        city_lat = city_response['coord']['lat']
        city_temp = (city_response['main']['temp'])+33.8
        city_humd = city_response['main']['humidity']
        city_winds = city_response['wind']['speed']
        city_clouds = city_response['clouds']['all']
    
   
        allcities_df.set_value(index, "Country",city_country)
        allcities_df.set_value(index, "Temparature(F)",city_temp)
        allcities_df.set_value(index, "Longitude", city_lon)
        allcities_df.set_value(index, "Latitude", city_lat)
        allcities_df.set_value(index, "Humidity", city_humd)
        allcities_df.set_value(index, "Cloudiness", city_clouds)
        allcities_df.set_value(index, "Wind Speed", city_winds)
    except(KeyError, IndexError):
        print("NaN")
    
allcities_df
```

```python
allcities_df = allcities_df.dropna(how = 'any')
allcities_df
```

```python
# Build a scatter plot for each data type
plt.scatter(allcities_df["Temparature(F)"].astype(float),
            allcities_df["Latitude"].astype(float),
            edgecolor="blue", linewidths=1, marker="o",
            alpha=0.8, label="City Name")
# Incorporate the other graph properties
plt.title("Temparature Vs. Latitude on 3/18/18")
plt.ylabel("Max.Temparature")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-20, 80])
plt.ylim([-50, 75])

# Save the figure
plt.savefig("Latitude Vs Temparature Plot.png")

# Show plot
plt.show() 
```
